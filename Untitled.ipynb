```python
results = pd.read_csv('resultados/results.csv').iloc[:, 1:]
i = 0
while i < results.shape[0]:
    if results.iloc[i, 3] != 1:
        results.iloc[i, 3] = str(int((results.iloc[i, 3] / results.iloc[i, 1])*100)) + '%'
    
    i += 1


results['ObjVal_h2'] = results[['ObjVal_h', 'ObjVal_h2']].min(axis = 1)
results['ObjVal'] = results[['ObjVal', 'ObjVal_h2']].min(axis = 1)
results['Runtime_max'] = 3600
results['Runtime'] = results[['Runtime', 'Runtime_max']].min(axis = 1)
del(results['Runtime_max'])
results['Gap'] = 100*results['Gap']
results['Gap_h'] = 100*(results['ObjVal_h'] - results['ObjVal']) / results['ObjVal_h']
results['Gap_h2'] = 100*(results['ObjVal_h2'] - results['ObjVal']) / results['ObjVal_h2']
del(results['ObjVal_h'])
del(results['ObjVal_h2'])
i = 0
while i < results.shape[0]:
    if results.iloc[i]['A4'] == False:
        results.iloc[i, 2] = 10
        results.iloc[i+1, 2] = 20
        results.iloc[i+2, 2] = 50
        i += 3
    else:
        results.iloc[i, 2] = 100
        i += 1

i = 0

print(results)

results.to_csv('resultados/results_modificados.csv')

tabla_comparadora = results.groupby(['n_N', 'n_B', 'A4', 'Init', 'k']).mean()[['Runtime', 'ObjVal', 'Gap', 'Runtime_h', 'Runtime_h2', 'Gap_h', 'Gap_h2']].round(2).reset_index()
tabla_comparadora.to_excel('resultados/summary_results_new2.xlsx')
```

```python
results1 = results[results['Init'] == True].copy()
results1['Gap'] = results1['Gap'].fillna(100)
results1['Runtime'] = results1['Runtime'].fillna(3600)

# for i in range(results1.shape[0]):
#     results1.iloc[i, 12] = results2.iloc[i, 12]
#     results1.iloc[i, 13] = results2.iloc[i, 13]

#     results1.iloc[i, 14] = results2.iloc[i, 14]
#     results1.iloc[i, 15] = results2.iloc[i, 15]

results1false = results[results['Init'] == False].copy()

results = pd.concat([results1, results1false])

# results1.to_csv('resultados_buenos5.csv')
tabla_comparadora = results.groupby(['n_N', 'n_B', 'A4', 'Init', 'k']).describe()[['Runtime', 'ObjVal', 'Gap', 'Runtime_h', 'Runtime_h2', 'Gap_h', 'Gap_h2']].round(2).reset_index()
# tabla_comparadora.to_excel('summary_results_new.xlsx')

tabla_comparadora
```

```python
resultados = pd.DataFrame()

resultados['n_N'] = tabla_comparadora['n_N']
resultados['A4'] = tabla_comparadora['A4']
resultados['n_B'] = tabla_comparadora['n_B']
resultados['k'] = tabla_comparadora['k']
resultados['Initialization'] = tabla_comparadora['Init']
resultados['#Found'] = tabla_comparadora['Gap']['count'].astype(int)
resultados['Gap'] = tabla_comparadora['Gap']['mean']
resultados['Runtime'] = tabla_comparadora['Runtime']['mean']
resultados['Gap_math'] = tabla_comparadora['Gap_h']['mean']
resultados['Runtime_math'] = tabla_comparadora['Runtime_h']['mean']
resultados['Gap_build'] = tabla_comparadora['Gap_h2']['mean']
resultados['Runtime_build'] = tabla_comparadora['Runtime_h2']['mean']


resultados.to_excel('summary_results_new2.xlsx')
```
