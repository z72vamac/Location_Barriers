```python
import matplotlib.pyplot as plt

for j in range(9, 10):
    file = pd.read_csv('instances_casestudy/zones/zone'+str(j)+'.txt', header=None, sep = "\t")
    file = file.rename(columns={0: "x_coord", 1: "y_coord", 2: "zone", 3: "building"})
    file.y_coord = 200 - file.y_coord
    for i in np.unique(file.building).tolist():
        building1 = file[file.building == i]
        plt.plot(building1.x_coord, building1.y_coord)
```

```python

fig, ax = plt.subplots()


barriers = []

for j in range(2, 3):

    file = pd.read_csv('zone'+str(j)+'.txt', header=None, sep = "\t")
    file = file.rename(columns={0: "x_coord", 1: "y_coord", 2: "zone", 3: "building"})
    file.y_coord = 200 - file.y_coord

    for build in np.unique(file.building).tolist():
        building = file[file.building == build]

        for i in range(building.shape[0]-1):
            barriers.append([building.x_coord.iloc[i], building.y_coord.iloc[i], building.x_coord.iloc[i+1], building.y_coord.iloc[i+1]])

        barriers.append([building.x_coord.iloc[i+1], building.y_coord.iloc[i+1], building.x_coord.iloc[0], building.y_coord.iloc[0]])
    
    # np.savetxt('barriers{0}.csv'.format(j), barriers, delimiter = ",")

for b in barriers:
    ax.plot([b[0], b[2]], [b[1], b[3]], c="red")
```
