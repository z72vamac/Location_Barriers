```python
import pandas as pd
import numpy as np
```

```python
results = pd.read_csv("resultados/resultados_buenos.csv").iloc[:,1:]
```

```python
print(results)
```

```python
results[['Instance', 'n_N', 'n_B', 'k', 'wL']] = results[['Instance', 'n_N', 'n_B', 'k', 'wL']].apply(np.int64)
```

```python
tabla_comparadora = results.groupby(['n_N', 'A4', 'Init', 'k']).describe()[['Runtime', 'ObjVal', 'Gap', 'Runtime_h', 'Runtime_h2', 'ObjVal_h', 'ObjVal_h2']].round(2).reset_index()
```

```python
tabla_comparadora.to_excel('resultados/summary_results.xlsx')
```

```python
results2 = pd.read_csv("resultados/resultados_buenos4.csv").iloc[:,1:]
```

```python
results1 = results[results['Init'] == True].copy()

for i in range(results1.shape[0]):
    results1.iloc[i]['ObjVal_h'] = results2.iloc[i]['ObjVal_h']
    results1.iloc[i]['Runtime_h'] = results2.iloc[i]['Runtime_h']
    
#     results1.iloc[i]['ObjVal_h2'] = results2.iloc[i]['ObjVal_h2']

results1.to_csv('resultados_buenos5.csv')
tabla_comparadora = results1.groupby(['n_N', 'A4', 'Init', 'k']).describe()[['Runtime', 'ObjVal', 'Gap', 'Runtime_h', 'Runtime_h2', 'ObjVal_h', 'ObjVal_h2']].round(2).reset_index()
tabla_comparadora.to_excel('summary_results_new.xlsx')
```
